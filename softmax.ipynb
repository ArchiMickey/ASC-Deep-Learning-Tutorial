```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

softmax = lambda x: np.exp(-x)/np.exp(-x).sum(-1)[:, :, np.newaxis]
x = np.linspace(-5, 5, 20)
X, Y, X2 = np.meshgrid(x, x, [0])
inputs = np.dstack((X, Y, X2))
prob = softmax(inputs)

ax.scatter(X.flatten(), Y.flatten(), prob[:, : , 0], c='blue')
ax.scatter(X.flatten(), Y.flatten(), prob[:, : , 1], c='red')
ax.scatter(X.flatten(), Y.flatten(), prob[:, : , 2], c='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Probability')
```
